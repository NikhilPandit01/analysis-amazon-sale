```python
data.drop(['New','PendingS',],axis=1 , inplace=True )
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data['ship-postal-code']=data['ship-postal-code'].astype('int')
data
```

```python
data['Date']=pd.to_datetime(data['Date'], format='%Y-%m-%d')
```

```python
data.rename(columns={'Qty':'Quantity'},inplace=True)
```

```python
data.describe(include=object)
```

#Exploratory Data Analysis

```python
ax=sns.countplot(x='Size',data=data)
for bars in ax.containers:
    ax.bar_label(bars)
plt.    
```

NOTE : frpm above graphs you can see that most of the people buys M- size

```python
data_x=data.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
data_x
```

```python
sns.barplot(x='Size',y='Quantity',data=data_x ,palette='Accent')
plt.show()
```

# NOTE: FROM ABOVE GRAPH YOU CAN SEE THAT MOST QTY SOLD OF M-SIZE

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Courier Status',data=data, hue='Status')
plt.show()
```

```python
sns.displot(data['Category'])
plt.xticks(rotation=90)
plt.title('category wise')
plt.show()
```

# From above graph we can say most of the buyers are t-shirt

```python
# Checking B2B Data  by using pie chart 
B2B_Check = data['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

Note : From above chart we can see that maximum i.e. 99.3% of buyers are retailers and 0.7% are B2B buyers

```python
sns.scatterplot(x='Category',y='Size',data=data)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x='ship-state',data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_state = data['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=data[data['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

#Note: above graph shows the top 10 buys state
